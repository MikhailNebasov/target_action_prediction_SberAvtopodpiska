2.3. Объединение таблиц.

Объединение таблиц выполняется по общему полю session_id. Однако для объединения не используется метод merge, поскольку не видится целесообразнмы получение таблицы большого объема с избыточностью. Вместо этого планируется проход по таблице sessions с переносом в нее фич только со значимыми аттрибутами. Т. е. выполняется проверка на наличие "целевого действия" в сессии. И если "да", то сохраняется наименьший номер hit_number в сессии, при котором было достигнуто "целевое действие" и сопутствующие ему данные. Если "нет", то наибольшее значение hit_number в сессии и также сопутствующие ему данные. В основе данного подхода лежит предположение, что чем дольше клиент взаимодействует с сервисом, тем больше вероятность достижения "целевого действия". И наоборот, клиенты не заинтересованные покидают сервис раньше.

```python
list(df_sessions.session_id)
```

```python
for element in list(df_sessions.session_id):
    print(df_hits[(df_hits.session_id == element) & (df_hits.event_action.isin([
        'sub_car_claim_click', 'sub_car_claim_submit_click', 'sub_open_dialog_click', 'sub_custom_question_submit_click',
        'sub_call_number_click', 'sub_callback_submit_click', 'sub_submit_success', 'sub_car_request_submit_click']))])
    break
```
